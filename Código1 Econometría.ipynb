### DATOS ESCOLARIDAD HOMBRES Y MUJERES

```python
intercensal_educación = download_data_from_mongoatlas('intercensal_2015', 'personas')
escolaridad = intercensal_educación[['clave_mun','Mujer Escolaridad Pob','Hombre Escolaridad Pob','Hombre Escolaridad','Mujer Escolaridad']]
def create__ecolaridad(df,column):
    años=[]
    for i in range (len(df)):
        v = df[column][i]
        h= df['Mujer Escolaridad Pob'][i]
        r = df['Hombre Escolaridad Pob'][i]
        if column == 'Mujer Escolaridad':
            escol= v/h
        else:
            escol= v/r
        años.append(escol)
    df[column + '_años_promedio'] = años
    return df

intercensal_educacion=create__ecolaridad(escolaridad,'Mujer Escolaridad')
intercensal_educacion=create__ecolaridad(escolaridad,'Hombre Escolaridad')
intercensal_educacion=intercensal_educacion[['clave_mun','Mujer Escolaridad_años_promedio','Hombre Escolaridad_años_promedio']]
intercensal_educacion
```

### DATOS CASOS FEMINICIDIO 

```python
snsp_mpos = download_data_from_mongoatlas('sesnsp', 'municipal_carpetas')
def extract_df_between_dates(df, start_date, end_date):
    df_fechas = df.loc[(df['date'] >= start_date) & (df['date'] <= end_date)]
    df_fechas = df_fechas.sort_values(['date', 'clave_mun'], ascending=True)
    df_fechas = df_fechas.reset_index(drop=True)    
    return df_fechas
start_date = '2015/01'
end_date = '2020/03'
snsp_mpos_dates = extract_df_between_dates(snsp_mpos, start_date, end_date)
snsp_mpos_dates = snsp_mpos_dates[['clave_mun','pob','date','1FFF', '1FFB', '1FFE', '1FFN']]
for ind, row in snsp_mpos_dates.iterrows():
   snsp_mpos_dates.loc[ind,"Casos de feminicidio total"]= row['1FFF']+row['1FFB']+row['1FFE']+row['1FFN']
def create__tasas_100mil(df,column):
    tasas=[]
    for i in range (len(df)):
        v = df[column][i]
        p = df['pob'][i]
        v_tasa = v / (p / 100000)
        tasas.append(v_tasa)
    df[column + '_tasa'] = tasas
    return df
snsp_mpos_dates=create__tasas_100mil(snsp_mpos_dates,'Casos de feminicidio total')
snsp_mpos_dates= snsp_mpos_dates.groupby(['clave_mun']).sum()
snsp_mpos_dates
```
